```python
#CI                                                                                    OK
def ci(tipo, enable):
    global REM, CIinicial
    if(tipo == "instrucao"):
        if(enable == "y"): print("\nREM (",REM,") recebe CI (",CIinicial,").")
        REM = CIinicial
        if(enable == "y"):
            print("\nREM (",REM,").")
            print("\nCI(",CIinicial,") recebe mais 1.")
        CIinicial+=1
        if(enable == "y"): print("\nCI(",CIinicial,").")
```

```python
#REM
def rem(REM, enable):
    global tipo
    if(REM>=0 and REM<EndinicialInst):
        tipo = "dado"
    if(REM>=EndinicialInst and REM<=EndfinalInst):
        tipo = "instrucao"
    if(enable == "y"): print("\nO endereço no REM é um(a)",tipo)
```

```python
#UC
#Memória de instruções
#Instruções:
#0 | HLT --> Finaliza o programa                                                     OK
def HLT(CIfinal, enable):                                                                   
    global CIinicial
    CIinicial = CIfinal
    if(enable == "y"): print("\nO programa é finalizado.")
#1 | LDA Op --> ler (op) no R0 (armazena)                                            OK                      
def LDA(RDM, enable):
    #instrucao
    global R0
    if(enable == "y"): print("\nR0 (",R0,") recebe RDM (",RDM,").")
    R0 = RDM
    if(enable == "y"): print("\nR0 (",R0,").")
#2 | STR Op --> escreve R0 no endereço (op)                                         OK    
def STR(R0, operando, enable):
    global REM, RDM
    if(enable == "y"):
        print("\nInstrução foi para a ULA.")
        print("\nREM (",REM,") recebe endereço do operando (",operando,").")
    REM = int(operando,2)
    if(enable == "y"):
        print("\nREM(",REM,")")
        print("\nRDM (",RDM,") recebe R0 (",R0,").")
    RDM = R0
    if(enable == "y"):
        print("\nRDM(",RDM,")")
        print("\nEndereço da memória principal (",REM,") recebe RDM (",RDM,").")
    mdadosmp[REM] = RDM 
    if(enable == "y"): print("\nCélula da memória desse endereço (",mdadosmp[REM],")")
#3 | ADD Op --> soma (op) no R0                                                      OK
def ADD(RDM, enable):
    global R0
    if(enable == "y"): print("\nInstrução foi para a ULA.")
    soma = int(RDM,2) + int(R0,2)
    if(enable == "y"): print("\nÉ somado RDM (",RDM,") em R0 (",R0,").")
    R0 = format(soma,'012b')
    if(enable == "y"): print("\nR0 (",R0,").")
#4 | SUB Op --> subtrai (op) de Ro                                                    OK
def SUB(RDM, enable):
    global R0
    if(enable == "y"): print("\nInstrução foi para a ULA.")
    subtracao = int(R0,2) - int(RDM,2)
    if(enable == "y"): print("\nÉ subtraído RDM (",RDM,") em R0 (",R0,").")
    R0 = format(subtracao,'012b')
    if(enable == "y"): print("\nR0 (",R0,").")
#5 | JZ Op --> pula se R0 = 0                                                         OK
def JZ(R0, operando, enable):
    global CIinicial, tipo
    if(enable == "y"): print("\nInstrução foi para a ULA.")
    if(int(R0,2) == 0):
        if(enable == "y"): print("\nCI (",CIinicial,") recebe operando (",operando,").")
        CIinicial = int(operando,2)
        if(enable == "y"): print("\nCI (",CIinicial,").")
        tipo = "instrucao"
#6 | JP Op --> se R0 > 0                                                              OK
def JP(R0, operando, enable):  
    global CIinicial, tipo
    if(enable == "y"): print("\nInstrução foi para a ULA.")
    if(int(R0,2) > 0):
        if(enable == "y"): print("\nCI (",CIinicial,") recebe operando (",operando,").")
        CIinicial = int(operando,2)
        if(enable == "y"): print("\nCI (",CIinicial,").")
        tipo = "instrucao"
#7 | JN Op --> se R0 < 0                                                              OK
def JN(R0, operando, enable):  
    global CIinicial, tipo
    if(enable == "y"): print("\nInstrução foi para a ULA.")
    if(int(R0,2) < 0):
        if(enable == "y"): print("\nCI (",CIinicial,") recebe operando (",operando,").")
        CIinicial = int(operando,2)
        if(enable == "y"): print("\nCI (",CIinicial,").")
        tipo = "instrucao"
#8 | JMP Op --> pula independente do R0                                               OK
def JMP(operando, enable):
    global CIinicial, tipo
    if(enable == "y"):
        print("\nInstrução foi para a ULA.")
        print("\nCI (",CIinicial,") recebe operando (",operando,").")
    CIinicial = int(operando,2)
    if(enable == "y"): print("\nCI (",CIinicial,").")
    tipo = "instrucao"
#9 | GET Op --> ler dado da porta de entrada e armazena no (op)                       OK
def GET(operando):
    print("#"*20)
    mdadosmp[int(operando,2)] = input("\nPorta de entrada: ")
    print("#"*20)
#A | PRT Op --> coloca na porta referente à impressora (op) o valor armazenado        OK
def PRT(operando):
    print("#"*20)
    print("\nPorta de impressão: ",mdadosmp[int(operando,2)],"\n")
    print("#"*20)
```

```python
#Decodificador de instruções
def decinst(instrucao, enable):
    global REM, RDM
    if(int(instrucao,2)==0):
        #leitura
        #dados
        HLT(CIfinal, enable)
    if(int(instrucao,2)==1):
        #leitura
        #busca de operando
        if(enable == "y"): print("\nREM (",REM,") recebe endereço do operando (",operando,").")
        REM = int(operando,2)
        if(enable == "y"): 
            print("\nREM (",REM,").")
            print("\nRDM (",RDM,") recebe informação do endereço (",REM,") da memória principal.")
        RDM = mdadosmp[REM]
        if(enable == "y"): print("\nRDM (",RDM,").")
        #dados
        LDA(RDM, enable)
    if(int(instrucao,2)==2):
        #escrita
        #busca de operando
        if(enable == "y"): print("\nREM (",REM,") recebe endereço do operando (",operando,").")
        REM = int(operando,2)
        if(enable == "y"): 
            print("\nREM (",REM,").")
            print("\nRDM (",RDM,") recebe informação do endereço (",REM,") da memória principal.")
        RDM = mdadosmp[REM]
        if(enable == "y"): print("\nRDM (",RDM,").")
        #dados
        #dados
        STR(R0, operando, enable)
    if(int(instrucao,2)==3):
        #leitura
        #busca de operando
        if(enable == "y"): print("\nREM (",REM,") recebe endereço do operando (",operando,").")
        REM = int(operando,2)
        if(enable == "y"):
            print("\nREM (",REM,").")
            print("\nRDM (",RDM,") recebe informação do endereço (",REM,") da memória principal.")
        RDM = mdadosmp[REM]
        if(enable == "y"): print("\nRDM (",RDM,").")
        #dados
        #dados
        ADD(RDM, enable)
    if(int(instrucao,2)==4):
        #leitura
        #busca de operando
        if(enable == "y"): print("\nREM (",REM,") recebe endereço do operando (",operando,").")
        REM = int(operando,2)
        if(enable == "y"): 
            print("\nREM (",REM,").")
            print("\nRDM (",RDM,") recebe informação do endereço (",REM,") da memória principal.")
        RDM = mdadosmp[REM]
        if(enable == "y"): print("\nRDM (",RDM,").")
        #dados
        #dados
        SUB(RDM, enable)
    if(int(instrucao,2)==5):
        #leitura
        #instrução
        JZ(R0, operando, enable)
    if(int(instrucao,2)==6):
        #leitura
        #instrução
        JP(R0, operando, enable)
    if(int(instrucao,2)==7):
        #leitura
        #instrução
        JN(R0, operando, enable)
    if(int(instrucao,2)==8):
        #leitura
        #instrução
        JMP(operando, enable)
    if(int(instrucao,2)==9):
        #leitura
        #dados
        GET(operando)
    if(int(instrucao,2)==10):
        #leitura
        #dados
        PRT(operando)
```
