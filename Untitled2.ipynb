1) Uma memória cache associativa por conjunto consiste em 256 linhas divididas em conjuntos de duas linhas cada. À memória principal contém 4K blocos de 128 bytes cada um. Mostre o formato de um endereço de MP e da cache.

```python
import math
qtdlinhas = 256
linhaporconjunto = 2
qtdbloco = 2**12
M = 8
largurabloco = 128

N = qtdbloco* largurabloco
T =  N * M
EMP = math.log2(N)
print("E:",int(EMP),"bits")


campCELULA = math.log2(largurabloco)
print("Célula:",int(campCELULA),"bits")

campCON = math.log2(qtdlinhas/linhaporconjunto)
print("Conjunto:",int(campCON),"bits")

campTAG = EMP-campCON-campCELULA
print("TAG:",int(campTAG),"bits")
```

2) Em que circunstâncias uma cache que funciona com mapeamento associativo por conjunto pode ser
considerada igual à cache que funciona com mapeamento direto?

R: O método de mapeamento associativo por conjunto é conhecido como o meio termo entre mapeamento direto e mapeamento associativo. Sua relação com o mapeamento direto está no fato de que ele apresenta endereços fixos (diretos) para os conjuntos, de maneira análoga ao que o direto tem com as linhas.

3) Considere um sistema de computação que possui uma memória principal (RAM) com capacidade máxima
de endereçamento de 64K células, e que cada célula armazena um byte de informação Para criar um
Sistema de controle e funcionamento da sua memória cache, a memória RAM é constituída de blocos de
8 bytes cada. À memória cache do sistema é do tipo mapeamento direto, contendo 32 linhas. Pergunta-se:

a) Como seria organizado o endereço da MP (RAM) em termos de tag, número da linha e do byte dentro de uma linha?

b) Em que linha estaria contido o byte armazenado no seguinte endereço da MP: 0001 0001 0001 1011?

c) Qual é a capacidade da memória cache em bytes?



```python
M = 8
capaRAM = 2**16
qtdlinhas = 32
larguradobloco = 8*M


print("A)")
#A
E = math.log2(capaRAM)
print("E:",int(E),"bits")

campLinha = math.log2(qtdlinhas)
print("linha:",int(campLinha),"bits")

campCELULA = math.log2(larguradobloco/M)
print("célula:",int(campCELULA),"bits")

campTAG = E-campLinha-campCELULA
print("TAG:",int(campTAG),"bits")

#B

print("\nB)")
x = "0001000100011011"[::-1]
lin=""
for i in range(int(campLinha)):
    lin += x[i+int(campCELULA)]
lin = lin[::-1]
print("Endereço de linha:",lin)

#C

print("\nC)")
#Tc = Tlinhas*Tcelulas*TM
TC = qtdlinhas*2**campCELULA
print("Capacidade da cache:",int(TC),"bytes")


```

6) Ao se verificar a organização de uma memória cache e da MP de um sistema de computação, observa-
se que a cache tem uma capacidade de armazenamento muito menor que a MP e, ainda assim, sabe-se
que em 100 acessos do processador à cache ele obtém cerca de 95 a 98% de acertos. Por quê?

R: Primeiramente por causa do preço, uma vez que a tecnologia da cache tem um custo muito mais alto que da MP. Em segundo lugar, por conta do próprio objetivo da cache, de ser uma memória intermediária entre a MP e o processador, armazenando somente o necessário, não toda a informação, aumentando assima velocidade das operações. 
    Não há como ter 100% de acertos porque uma hora a cache terá que ir buscar dados que faltam na memória principal, ocorrendo, portanto, um miss.

7) Por que é necessário se estabelecer uma política para substituição de linhas para os mérodos de mapeamento associativo e não para o método direto?

R: No método de mapeamento direto cada bloco está permanentemente associado a uma linha da cache, enquanto no método associativo isso não ocorre, e nesse caso, é preciso decidir qual bloco será retirado para dar vez ao bloco que está sendo buscado



8) Considere uma MP que possui 4K blocos de 128 células de 1 byte cada e uma memória cache do tipo
associativa por conjunto que possui 64 linhas divididas em conjuntos de quatro linhas. Qual deverá ser
o formato do campo de endereçamento para acesso à MP e à memória cache?


```python
import math
qtdblocos = 4*2**10
celulaporbloco = 128
M = 8
qtdlinhas = 64
largconj = 4
capaMP = qtdblocos*celulaporbloco*M
qtdconjuntos = qtdlinhas/largconj
#formato do campo de endereçamento para acesso à MP e à cache?
E = math.log2(celulaporbloco*qtdblocos)
print("E:",E,"bits")
campCONJ = math.log2(qtdconjuntos)
print("Conjunto:",campCONJ,"bits")
campCEL = math.log2(celulaporbloco)
print("Célula:",campCEL,"bits")
campTAG = E - campCEL - campCONJ
print("TAG:",campTAG,"bits")
```

reço hexadecimal: SD7A9F2. Sabendo-se que neste sistema cada linha da cache tem atribuídos 512

b) Qual foi a linha acessada pelo processador?
